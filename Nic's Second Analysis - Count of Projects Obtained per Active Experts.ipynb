```python
#import library psycopyg2
import psycopg2

#import library pandas
import pandas as pd

#import library sqlio
import pandas.io.sql as sqlio

#import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#create database connection variable 
conn = psycopg2.connect(user="rukwz22tz3nz8al",password="OMKSeKL5qE62X9P_4px7gdQM2mnmS-pjM0n7Nc0Z",host="evening-soiree.sequindb.com",database="dbj7awza4t09pvi")

#define query
query1 = """
SELECT "source 2"."No of Projects", count(*) as "Experts Count" FROM
(SELECT count(*) as "No of Projects", "source"."Name" FROM 
(SELECT selectsystems.record_id, experts.name as "Name" 
FROM experts LEFT JOIN selectsystems 
ON selectsystems.expert = experts.id 
WHERE selectsystems.record_id IS NOT NULL)
"source" GROUP BY "source"."Name")
"source 2" 
GROUP BY "source 2"."No of Projects" 
ORDER BY "Experts Count" DESC
"""

#execute query and save it to a variable
dataset1 = sqlio.read_sql_query(query1,conn)
df = pd.DataFrame(data=dataset1)
xdata=df['No of Projects']
ydata=df['Experts Count']
plt.figure(figsize=(15, 7))
plt.bar(xdata,ydata,color ='maroon',width = 0.98)
plt.xlabel("No of Projects per Expert")
plt.ylabel("No of Experts")
plt.title("How Many Projects Do Each Expert Acquire? - Data Analysis Prepared by Nic Rayce")
ax = plt.subplot()
ax.grid(color='grey', linestyle='dashed', linewidth=0.5)
plt.show()

```
